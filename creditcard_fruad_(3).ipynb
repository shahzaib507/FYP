```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt


# Ensure Target Variable is Numeric
# Check unique values in 'Class' column
print("Unique values in Class before conversion:", df['Class'].unique())
df['Class'] = df['Class'].apply(lambda x: int(str(x).replace("b'", "").replace("'", "")))
print("Unique values in Class after conversion:", df['Class'].unique())

# Feature Engineering
df['Time_of_Day'] = df['Time'] % (24 * 60 * 60)  # Time modulo seconds in a day
df['Hour'] = df['Time_of_Day'] // 3600          # Convert to hours
df['Log_Amount'] = np.log1p(df['Amount'])       # Log transform for skewed 'Amount'

# Feature Scaling
scaler = StandardScaler()
scaled_columns = ['Time', 'Log_Amount']  # Specify columns to scale
df[scaled_columns] = scaler.fit_transform(df[scaled_columns])

# Handle Imbalance (Separate Features and Target)
X = df.drop(columns=['Class', 'Amount', 'Time_of_Day'])  # Exclude redundant columns
y = df['Class']
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42
)

# Apply Random Forest
model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

# Feature Importance
importances = model.feature_importances_
important_features = pd.Series(importances, index=X_train.columns).sort_values(ascending=False)

# Display Feature Importance
print("Feature Importance:\n", important_features)

# Optional: Plot Feature Importance
plt.figure(figsize=(10, 6))
important_features.plot(kind='bar')
plt.title('Feature Importance from Random Forest')
plt.ylabel('Importance Score')
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, f1_score

# Predictions
y_pred = model.predict(X_test)

# Metrics Calculation
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print Metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# Detailed Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Check and Convert `y` to Binary Format
mlb = MultiLabelBinarizer()
y_binary = mlb.fit_transform(y)

# Step 2: Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y_binary, test_size=0.3, random_state=42)

# Step 3: Standardize Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Step 4: Train Multi-label SVM
svm_model = OneVsRestClassifier(SVC(kernel='rbf', C=1, gamma='scale', random_state=42))
svm_model.fit(x_train, y_train)

# Step 5: Predictions and Evaluation
y_pred = svm_model.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix (One per label):")
for i, label in enumerate(mlb.classes_):
    print(f"\nLabel {label}:")
    print(confusion_matrix(y_test[:, i], y_pred[:, i]))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier

# Assuming `df` is already loaded as a DataFrame
# Step 1: Splitting Features (x) and Target (y)
x = df.iloc[:, :-1]
y = df.iloc[:, len(df.columns) - 1]

# Select specific columns for `x` as per ncols
ncols = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
         'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
         'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
x = x[ncols]

# Check shapes
print("x.shape:", x.shape)
print("y.shape:", y.shape)
# Ensure target variable y is of integer type
y = y.astype(int)
# Step 2: Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# Step 3: Train the XGBoost Model
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)
xgb_model.fit(x_train, y_train)

# Step 4: Predictions and Evaluation
y_pred = xgb_model.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Step 5: Calculate Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score: {accuracy:.2f}")
```

**Using Smot XGBoost Classifier**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE



# Step 1: Splitting Features (x) and Target (y)
x = df.iloc[:, :-1]  # Select all columns except the last as features
y = df.iloc[:, -1]   # Select the last column as the target variable

# Ensure target variable y is of integer type
y = y.astype(int)

# Select specific columns for `x` (optional, based on the dataset)
ncols = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
         'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
         'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
x = x[ncols]

# Step 2: Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42, stratify=y
)

# Step 3: Handle Class Imbalance with SMOTE
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)

# Check class distribution after SMOTE
print("Class distribution before SMOTE:", y_train.value_counts())
print("Class distribution after SMOTE:", pd.Series(y_train_smote).value_counts())

# Step 4: Train the XGBoost Model
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)
xgb_model.fit(x_train_smote, y_train_smote)

# Step 5: Predictions and Evaluation
y_pred = xgb_model.predict(x_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Step 6: Calculate Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score: {accuracy:.2f}")
```
