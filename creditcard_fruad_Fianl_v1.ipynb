```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Check and Convert `y` to Binary Format
mlb = MultiLabelBinarizer()
y_binary = mlb.fit_transform(y)

# Step 2: Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y_binary, test_size=0.3, random_state=42)

# Step 3: Standardize Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Step 4: Train Multi-label SVM
svm_model = OneVsRestClassifier(SVC(kernel='rbf', C=1, gamma='scale', random_state=42))
svm_model.fit(x_train, y_train)

# Step 5: Predictions and Evaluation
y_pred = svm_model.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix (One per label):")
for i, label in enumerate(mlb.classes_):
    print(f"\nLabel {label}:")
    print(confusion_matrix(y_test[:, i], y_pred[:, i]))
```

**Apply the XGB Classifier**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier

# Assuming `df` is already loaded as a DataFrame
# Step 1: Splitting Features (x) and Target (y)
x = df.iloc[:, :-1]
y = df.iloc[:, len(df.columns) - 1]

# Select specific columns for `x` as per ncols
ncols = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
         'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
         'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
x = x[ncols]

# Check shapes
print("x.shape:", x.shape)
print("y.shape:", y.shape)
# Ensure target variable y is of integer type
y = y.astype(int)
# Step 2: Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# Step 3: Train the XGBoost Model
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)
xgb_model.fit(x_train, y_train)

# Step 4: Predictions and Evaluation
y_pred = xgb_model.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Step 5: Calculate Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score: {accuracy:.2f}")
```
